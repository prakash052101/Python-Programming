```python
import os
import pandas as pd
import numpy as np
CancerData = pd.read_csv('Cancer.csv')
CancerData
```

```python
CancerData1 = CancerData.iloc[:, 1:32]
CancerData1['diagnosis'] = CancerData1['diagnosis'].map({'M':1, 'B':0})
CancerData1
```

```python
# step-3 - splitting to training and test data
from sklearn.model_selection import train_test_split
x = CancerData1.iloc[:, 1:32]
y = CancerData1['diagnosis']
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = .33)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#step 4 - Running Logistic regression
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
Model_cancer = lm.fit(x_train, y_train)
```

```python
Predicted_Y = Model_cancer.predict(x_test)
Predicted_Y
```

```python
x_test1 = x_test.reset_index()
y_test1 = y_test.reset_index()
```

```python
x_test1
```

```python
Predicted_Y1 = pd.DataFrame(Predicted_Y, columns = ["Predicted_Y"])
Predicted_Y1
```

```python
Test_with_Predict = pd.concat([x_test1, y_test1, Predicted_Y1], axis = 1)
Test_with_Predict
```

```python
# setp 6- checking Accuracy
pd.crosstab(Test_with_Predict.diagnosis, Test_with_Predict.Predicted_Y)
```

```python
# OR
from sklearn.metrics import confusion_matrix
confusion_matrix(Test_with_Predict.diagnosis, Test_with_Predict.Predicted_Y)
```

```python
from sklearn import metrics
metrics.accuracy_score(Test_with_Predict.diagnosis, Test_with_Predict.Predicted_Y)
```
