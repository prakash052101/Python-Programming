```python
import pandas as pd
Boston = pd.read_csv('boston.csv')
Boston
```

```python
x = Boston.iloc[:, 0:13]
y =  Boston['medv']
print(x.shape)
print(y.shape)
```

```python
x
```

```python
y
```

```python
import statsmodels.api as sm
x_with_constant = sm.add_constant(x)
x_with_constant
```

```python
results = sm.OLS(y, x_with_constant).fit()
print(results.summary())
```

```python
# drop insignificant variable
x_with_constant = x_with_constant.drop(['indus', 'age'], axis = 1)
results = sm.OLS(y, x_with_constant).fit()
print(results.summary())
```

```python
# making pridictions
pred_medv = results.predict(x_with_constant)
pred_medv
```

```python
# cancatinating
Predicted_boston = pd.concat([x_with_constant, y,pd.DataFrame(pred_medv, columns=['Predicted_Price'])], axis = 1)
Predicted_boston
```

```python
# Finding accuracy by each observation
Predicted_boston["Errors"] = abs(Predicted_boston['medv']-Predicted_boston['Predicted_Price'])/Predicted_boston['medv']
Predicted_boston
```

```python
# Finding accuracy of overall model
1 - Predicted_boston.Errors.mean()
```
